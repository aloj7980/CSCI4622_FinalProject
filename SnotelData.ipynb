```python
# make a dictionary of our stations individual dataframes
stations = {station: df[df['station_id'] == station] for station in df.station_id.unique()}
print(f'Found {len(stations)} distinct stations.')
```

```python
# split each of those into years
for s in stations:
    df = stations[s]
    df['date']= pd.to_datetime(df['date'],format='%Y-%m-%d')
    stations[s] = {year: df[df['date'].dt.year == year] for year in range(2013, 2023)}
```
